```python
age = 37
print("Tôi năm may " + str(age))

#f-string (ép kiểu)
print(f"Tôi năm nay {age}")

#kiểu ép chuỗi khác
print("Tôi năm nay {}".format(age))

#set PYTHONIOENCODING=utf8 
```

```python
# In hoa toàn bộ ký tự 
s = "hello"
s = s.upper()
print(s)
```

```python
# In hoa chữ cái đầu 
s = "hello"
# s = s.capitalize() 
s = s.title() #in hoa chữ cái đầu tiên từng từ trong câu 
print(s)
```
